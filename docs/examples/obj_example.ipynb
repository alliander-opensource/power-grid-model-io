```python
from power_grid_model_io.obj import InputGrid, UpdateGrid, OutputGrid
from power_grid_model import PowerGridModel
import pandas as pd
```

```python
input_data = InputGrid()

input_data.node.initialize(3)
input_data.node.id = [1, 2, 3]
input_data.node.u_rated = [10.5]

input_data.line.initialize(2)
input_data.line.id = [4, 5]
input_data.line.from_node = [1, 2]
input_data.line.to_node = [2, 3]


model = PowerGridModel(input_data)

output_data = OutputGrid(model.calculate_power_flow())

display(output_data.node.as_df())
```

```python
update_data = UpdateGrid()

update_data.line.initialize([1, 2, 1])
update_data.line[0]["id"] = [4]
update_data.line[1]["id"] = [4, 5]
update_data.line[2]["id"] = [5]
update_data.line[0]["from_status"] = [0]
update_data.line[1]["from_status"] = [0, 0]
update_data.line[2]["from_status"] = [0]

display(pd.DataFrame(update_data.line[0]))
display(pd.DataFrame(update_data.line[1]))
display(pd.DataFrame(update_data.line[2]))

output_data = OutputGrid(model.calculate_power_flow(update_data=update_data))

display(pd.DataFrame(output_data.node[0]))
display(pd.DataFrame(output_data.node[1]))
display(pd.DataFrame(output_data.node[2]))
```
